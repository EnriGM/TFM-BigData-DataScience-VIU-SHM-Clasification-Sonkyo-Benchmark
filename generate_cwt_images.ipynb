```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
import matplotlib as mpl
mpl.use('agg')
# paper de singapour usan morlet wavelet
```

```python
folder = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\data\a8_npy"
img_folder = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\cwt\cwt_no_comon_scale"

for file in os.listdir(folder):
    if "Case_" in file:
        print(datetime.now())
        print(file)
        data = np.load(os.path.join(folder, file))
        for count, signal in enumerate(data):
            dt = 1/len(signal)
            fs = len(signal)
            frequencies = np.array([400, 1]) / fs # the recording is 1666 hrz and the shaker uses 400 hrz so for both reasons is a good compromise
            scale = pywt.frequency2scale('morl', frequencies)
            scales = np.geomspace(scale[0], scale[-1], num=200)
            cwtmatr, freqs = pywt.cwt(signal, scales, "morl")
            cwtmatr = np.abs(cwtmatr[:-1, :-1])
            time = np.linspace(0, 1, len(signal))
            img_name = file.split(".")[0] + str(count).zfill(4) + ".png"
            dest_p = os.path.join(img_folder, img_name)
    
            fig= plt.figure(figsize=(1, 1), dpi=224)
            ax = plt.Axes(fig, [0., 0., 1., 1.])
            pcm = ax.pcolormesh(time, freqs, cwtmatr, cmap="jet")
            ax.set_yscale("log")
            ax.axis('off')
            fig.add_axes(ax)
            fig.savefig(dest_p, bbox_inches='tight', pad_inches=0.0)
            plt.close(fig)
            
```

```python
import pandas as pd
import os
import re

img_folder = r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\cwt\cwt_no_comon_scale"
index = list(range(len(os.listdir(img_folder))))

df = pd.DataFrame(index=index, columns=["filename", "case", "temperature"])

pattern = re.compile(r"\([^\)]*\)+")
for count, file in enumerate(os.listdir(img_folder)):
    df.loc[count, "filename"] = file
    df.loc[count,"case"] = file.split("_")[1]
    df.loc[count,"temperature"] = int(pattern.findall(file)[0].replace("(", "").replace(")", ""))

df.to_csv(r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\cwt\cwt_no_comon_scale_case_temp_info.csv")
```
