Random forest classifier time

```python
import numpy as np

x = np.load(r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\res\data_for_ml_all_std_force_quality\x_for_ml.npy")
y = np.load(r"C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\res\data_for_ml_all_std_force_quality\y_for_ml.npy")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
SEED = 42

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=SEED)

lb = LabelEncoder()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
from datetime import datetime

print(datetime.now())

n_iter_search = 10

param_dist = {"n_estimators":[10],
              "criterion":["gini", "entropy", "log_loss"],
              "min_samples_split":[2, 10 , 20],
              "min_samples_leaf": [1, 5, 8],
              "max_features": ["sqrt", "log2"]
              }

clf = RandomForestClassifier(random_state=SEED)
print("start random search")
random_search = RandomizedSearchCV(
    clf, param_distributions=param_dist, n_iter=n_iter_search,
    random_state=SEED, verbose=3)

random_search.fit(X_train, y_train)
print(random_search.best_params_)
model = random_search.best_estimator_
y_pred = model.predict(X_test)
print("precision", precision_score(y_test, y_pred, average="micro"))
print(datetime.now())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score
from sklearn.model_selection import GridSearchCV

param_dist = {"n_estimators":[100],
              "criterion":["entropy"],
              "min_samples_split":[50, 60, 70],
              "min_samples_leaf": [2],
              "max_features": ["sqrt"]
              }

clf = RandomForestClassifier(random_state=SEED)
print("start random search")
search = GridSearchCV(clf, param_grid=param_dist, verbose=3)

search.fit(X_train, y_train)
print(search.best_params_)
model = search.best_estimator_
y_pred = model.predict(X_test)
print("precision", precision_score(y_test, y_pred, average="micro"))
```

```python
import pickle 

with open(r'C:\Users\AndreaEnrique\Documents\TFM_MBID_VIU\models\time_all_std_force_quality\RFC_time.pickle', 'wb') as f:
    pickle.dump(model, f)
```
